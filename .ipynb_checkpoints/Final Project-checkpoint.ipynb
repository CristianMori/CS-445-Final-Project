```python
import cv2
import numpy as np
from scipy import linalg
from scipy import ndimage
from numpy.linalg import svd, inv
import scipy

%matplotlib inline
from matplotlib import pyplot as plt
```

# Text line tracing

```python
image = './input/sample1.jpg'
image = cv2.imread(image)
scale = 20
plt.figure(figsize = (scale*1.3333333333,scale))
plt.imshow(image)
```

## Calculate characteristic scale

```python
# Compute mean gradient magnitude of image
# http://campar.in.tum.de/Chair/HaukeHeibelGaussianDerivatives
def gaussianDerivative(image):
    sigma = 3
    derivative = ndimage.gaussian_filter(image, sigma, order=1)
    return derivative

# Calculate mean gradient magnitude
def MGM(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    derivative = gaussianDerivative(image)
    return np.mean(derivative)

def imagePyramid(image):
    numDownsamples = 10
    current = image
    MGMs = [MGM(image)]
    for i in range(numDownsamples):
        current = cv2.pyrDown(current)
        MGMs.append(MGM(current))
    return MGMs

def findPeak(MGMs):
    current = MGMs[0]
    for i in range(len(MGMs)):
        if MGMs[i] < current:
            return current, i
        current = MGMs[i]
    return current, len(MGMs) - 1

MGMs = imagePyramid(image)
plt.plot([i for i in range(11)], MGMs)
peak, characteristicScale = findPeak(MGMs)
```

## Initial line tracing

```python
def plotLines(image, lines):
    fig, ax = plt.subplots(figsize = (scale*1.3333333333,scale))
    ax.imshow(image, extent=[0, len(image[0]), 0, len(image)])
    for line in lines:
        ax.plot(line[1], line[0], '-', linewidth=2, color='firebrick')

def getPatch(image, x, y, patchSize):
    x0 = int(x - patchSize/2)
    y0 = int(y - patchSize/2)
    return image[y0: y0+patchSize, x0: x0+patchSize]

def randomPoint(corners):
    randInt = np.random.randint(0, high=len(corners))
    return corners[randInt]

def getCorners(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = np.float32(gray)
    blockSize = 2
    kSize = 3
    k = 0.04
    dst = cv2.cornerHarris(gray, blockSize, kSize, k)

    image[dst>0.01*dst.max()]=[255,255,255]
    image[dst<=0.01*dst.max()]=[0,0,0]
    plt.figure(figsize = (20*1.3333333333,20))
    plt.imshow(image)
    return np.argwhere(dst>0.01*dst.max())

def difference(patch1, patch2):
    return np.abs(np.sum(patch1) - np.sum(patch2))

def normalize(patch):
    sigma = np.std(patch)
    mean = np.mean(patch)
    return (patch - mean) / sigma

def findNextPatch(image, patch, x, y, steps, patchSize):
    bestCorrelation = 0
    bestX = 0
    bestY = 0
    for step in steps:
        x1 = x+step[0]
        y1 = y+step[1]
        patch1 = normalize(getPatch(image, x1, y1, patchSize))
        correlation = difference(patch, patch1)
        if correlation > bestCorrelation:
            bestCorrelation = correlation
            bestX = x1
            bestY = y1
    return bestX, bestY
    

def findLine(image, corners, s, angles1, patchSize):
    x, y = randomPoint(corners)
    line = []
    line.append([x])
    line.append([y])
    
    steps = [np.cos(angles1), np.sin(angles1)] * s
    patch = normalize(getPatch(image, x, y, patchSize))
    for asdf in range(1000): # Should be while in text region
        x, y = findNextPatch(image, patch, x, y, steps, patchSize)
        
        line[0].append(x)
        line[1].append(y)
        patch = normalize(getPatch(image, x, y, patchSize))
    return line
    
def lineTracing(image):
    corners = getCorners(image)
    
    original = image
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Downsample to characteristic scale
    for i in range(characteristicScale-1):
        image = cv2.pyrDown(image)
    scale = len(original)/ len(image)
        
    patchSize = 30
    s = 3 # Step size
    m = 10 # numOfAngles
    # Angles for going each direction
    angles1 = np.linspace(-np.pi/2, np.pi/2, num=m)
    angles2 = np.linspace(np.pi/2, 3*np.pi/2, num=m)
    
    # Find lines
    numLines = 1
    lines = []
    for i in range(numLines):
        lines.append(findLine(image, corners/scale, s, angles1, patchSize)*scale)
    return lines

lines = lineTracing(image)
plotLines(image, lines)
# getTextRegion(image)
```
